```python
from lib.BRV import random
import math
import matplotlib.pyplot as plt

count = 2000
boom = a = random(count=count)
b = [i for i in range(count)]
```

```python
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005
rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
plt.figure(figsize=(8, 8))

ax_scatter = plt.axes(rect_scatter)
ax_scatter.grid(which="major", linewidth=1.2, color="black")
ax_scatter.grid(which="minor", linestyle="--", color="green", linewidth=1)
ax_scatter.tick_params(which='major', length=10, width=2)
ax_histx = plt.axes(rect_histx)
ax_histx.grid(True)
ax_histx.tick_params(direction='in', labelbottom=False)

ax_scatter.scatter(a, b)
ax_histx.hist(a, normed=True)

ax_histx.set_xlim(ax_scatter.get_xlim())

plt.show()
```

```python
m = math.fsum(a) / count
print(".  : {}".format(m))
```

```python
sum = 0
for i in range(len(a)):
    sum += (a[i] - m) ** 2
d = sum / count
print(" : {}".format(d))
```

```python
s = 2
b = []
result = 0
for num, chislo in enumerate(list(boom)):
    if num + s < len(boom):
        b.append(chislo * a[num + s])
for i in b:
    result += i 

r = 1 / 12 * result / (count - s)
print(" : {}".format(r))


```
